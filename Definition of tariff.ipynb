# **Описание проекта:**

В нашем распоряжении данные популярного сервиса аренды самокатов GoFast. 

Предоставлены данные о некоторых пользователях из нескольких городов, а также об их поездках. Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. 

Сервисом можно пользоваться	без подписки и с подпиской `Ultra`.

# **Описание тарифов сервиса GoFast в мобильном приложении:**

- без подписки 

    - абонентская плата отсутствует;
    
    - стоимость одной минуты поездки — 8 рублей;
    
    - стоимость старта (начала поездки) — 50 рублей;
    
- с подпиской `Ultra`

    - абонентская плата — 199 рублей в месяц;
    
    - стоимость одной минуты поездки — 6 рублей;
    
    - стоимость старта — бесплатно.   

# **Описание данных:**

Для анализа были предоставлены три файла с данными: кто они - клиенты, откуда, каким типом подписки в мобильном приложении пользуются, тарифы по подписке, объемы ежемесячных платежей,  данные по поездкам.

Пользователи — `users_go.csv`

- `user_id`	уникальный идентификатор пользователя
- `name`	имя пользователя
- `age`	возраст
- `city`	город
- `subscription_type`	тип подписки (free, ultra)

Поездки — `rides_go.csv`
	
- `user_id`	уникальный идентификатор пользователя
- `distance`	расстояние, которое пользователь проехал в текущей сессии (в метрах)
- `duration`	продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»
- `date`	дата совершения поездки

Подписки — `subscriptions_go.csv`
	
- `subscription_type`	тип подписки
-` minute_price`	стоимость одной минуты поездки по данной подписке
- `start_ride_price`	стоимость начала поездки
- `subscription_fee`	стоимость ежемесячного платежа

### **Рассмотрим частоту встречаемости городов.**

*Для расчета частоты воспользуемся данными из датафрейма `users_go`*

*Продемонстрируем какие города представлены в этом датафрейме, сколько клиентов из этих городов пользуется сервисом ( то есть покажем частоту).*

*Данные по частоте получим из сводной таблицы.* 

*Для удобства анализа расчитаем, помимо абсолютной частоты, относительную частоту по каждому городу.*

*Для этого обьявим универсальный метод `percent_category`, который создаст сводную таблицу и позволит найти процент уникального значения столбца.*

```python
# метод создаст сводную таблицу, отсортирует и определит процент уникального значения столбца
def percent_category(new_df, index, column):
    try:
        sort_df = (
            new_df.pivot_table(index=index, values=column, aggfunc='count')
            .sort_values(by=column, ascending = False)
        )
        sort_df['percent'] = (sort_df[column] / sort_df[column].sum()) * 100  
        sort_df.columns = ['Количество', 'Процент от общего']
        return sort_df
    except:
        return 'Не верно введен аргумент'
```

```python
city_pivot = percent_category(users_go, 'city', 'user_id') # вызовем метод
city_pivot
```

### **Гипотеза 1. Пользователи с подпиской больше времени тратят на поездки, чем пользователи без подписки.**

***Первое задание:***
*Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверим гипотезу. Используем исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.*

***Первая гипотеза одностороняя для средних двух генеральных совокупностей***

*H0 (нулевая гипотеза) -  средняя продолжительность поездок пользователей с подпиской не отличается от средней продолжительности поездок пользователей без подписки.*

*H1 (альтернативная гипотеза) - средняя продолжительность поездок пользователей с подпиской больше средней продолжительности поездок пользователей без подписки.*

*Рассмотрим две независимые генеральные совокупности: `general_table_free` и `general_table_ultra`*

*Одним из условий применения т-теста является то, что выборочные средние нормально распределены.*

*Проверим выборки на аномалий, постороим график `boxplot` для обоих тарифов*

```python
general_table.boxplot(column='duration',  by='subscription_type', figsize=(6, 4))

plt.ylim(0, 41)
plt.title('Диаграмма размаха продолжительности одной поездки по тарифам')
plt.xlabel('Время')
plt.show()
```

*В выборке `free` выбросов нет, в выборке `ultra` есть выбросы.*

*Определим границы для выборки `ultra`.*

```python
duration_ultra= general_table_ultra.copy()

duration_ultra.boxplot(column='duration', figsize=(6, 4))
plt.ylim(3, 33)
plt.title('Диаграмма размаха продолжительности одной поездки тариф ultra')
plt.xlabel('Время')
plt.show()
```

*Границы согласно диаграммы размаха от 5 до 32 мин (включая крайние точки).*

*Устраним выбросы*

*Устранять выбросы будем на копии датафрейма `general_table_ultra`.*

*Применим условия:*

```python
# условия для времени тарифа ultra
duration_borders = "5 <= duration <= 32"
duration_ultra_cut =  duration_ultra.query(duration_borders)
duration_ultra_cut['duration'].describe()
```

*Количество объектов уменьшилось с 6500 до 6395. Среднее и медиана сравнялись по значениям*

*Т.к. выборки у нас независимые, воспользуемся функцией `ttest_ind` из модуля `scipy.stats` (так как у нас выборки разных размеров используем аргумет `equal_var=False`):*

```python
import scipy.stats as st

print(f'Среднее значение продолжительность поездок выборки ultra: {duration_ultra_cut.duration.mean() :.2f}')
print(f'Среднее значение продолжительность поездок выборки free: {general_table_free.duration.mean() :.2f}')

alpha = 0.05 # уровень статистической значимости.Хотим чтобы вероятность правильности нашего вывода была 95% и больше
print(f'Уровень статистической значимости: {alpha}')

# проведем стат. тест о равенстве средних значений времени двух выборок,
# в качестве альтернативной используем одностороннюю гипотезу "больше"
results = st.ttest_ind(duration_ultra_cut['duration'],
                       general_table_free['duration'],
                       alternative='greater',
                       equal_var=False)

print(f'Статистика t-теста: {results.statistic :.2f}')
print(f'Значение p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о том, что средняя продолжительность поездок\
 пользователей с подпиской не отличается от средней продолжительности поездок пользователей без подписки')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

***Вывод***

*Полученное значение `p-value` (~1.702360e-24) меньше заданного уровня значимости, поэтому можно отвергнуть нулевую гипотезу.*

*Есть основания говорить, что пользователи с подпиской тратят времени на поездки больше, чем пользователи без подписки.*

То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соотвествующий платёж. 
Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 10 % пользователей. Выясните, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %. Подберите параметры распределения, описывающего эту ситуацию, постройте график распределения и сформулируйте ответ на вопрос о количестве промокодов.*

*Начнем подбор с 1000 промокодов (где 100 подписок это 10% общего числа розданных промокодов)*

```python
from matplotlib import pyplot as plt
from scipy.stats import binom

n = 100 # количество попыток
p = 0.1 # вероятность успеха

distr = []

for k in range(0, n + 1):
    current_value = binom.pmf(k, n, p)
    distr.append(current_value)
    
plt.bar(range(0, n + 1), distr)
plt.show()
```

*График показывает, что при вероятности успеха p, равной 10%, и 1000 попытках вероятность, что случится больше 95  успехов (вероятность не выполнить план была бы примерно 5 %) ~40%.*

*Процент успехов мал. Будем наращивать количество промокодов.* 

```python
from math import factorial
def bin_prob(n1, k, m): # Формула Бернулли для подсчета функции распределения
    chose= factorial(n1)/(factorial(k)*factorial(n1-k)) # подсчёт числа сочетаний
    return chose * (m**k) * ((1 - m)**(n1 - k))

def cdf(k_maximum, n2, m): # функция распределения 
    value = 0
    for i in range(k_maximum + 1):
        value += bin_prob(n2, i, m)                        
    return value

k, m, p = [(float(i)) for i in input().split()] # ввод данных в одну строчку по условию:
# где к - нужное количество подписок (100 шт), m - вероятность продлят подписку (10%), 
#  р - вероятность не более 5% невыполнения плана (на 0.95% выполнить)

k_maximum = int(k)

n2 = 1000 # начинаем перебор с 1000 промокодов
while (1 - cdf(k_maximum, n2, m)) < p: # пока вероятность не станет удовлетворительной 
    n2 += 10 # Накидываем ещё по 10 шт количество разданных промокодов

print('Результат пересчета промокодов', int(n2)) 
```

*Ввод данных через пробел 100 0.1 0.95*

***Результат:***

*Необходимо раздать 1180 промокодов, что б получить 100 подписок с вероятностью не выполнить план была примерно 5 %*

Отдел планирует разослать 11 млн уведомлений. С помощью аппроксимации постройте примерный график распределения и оцените вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.*

```python
from math import sqrt
from scipy import stats as st

# задаём параметры биномиального распределения
n = 11000000
p = 0.4

# задаём параметры нормального распределения равными
# мат.ожиданию и дисперсии биномиального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение
distr = st.norm(mu, sigma) 

# считаем вероятность хвоста слева от 399.5 тысяч
result = distr.cdf(399500)
print(result)
```

*Вероятность того, что уведомления откроют не более 399,5 тыс. пользователей равна нулю.*

*Проверим правильность расчетов и подберем такое количество откликов с вероятностью 1%  - открыть меньше такого количества.*

*Вероятность того, что уведомления откроют не более 4 397  тыс. пользователей меньше 1%.*

```python
from scipy.stats import norm
 
# зададим функцию, которая выдаёт <num> чисел,
# равномерно распределённых от <start> до <stop>
# (понадобится для визуализации нормального распределения)
def linspace(start, stop, num):
    step = (stop - start) / (num - 1)
    result = []
    for i in range(num):
        result.append(start + step * i)
    return result
 
# задаём параметры биномиального распределения
n = 11000000
p = 0.4

# границы промежутка, в котором будем строить аппроксимацию -
# от n*p минус 4 ст.отклонения бином.распределения
# до n*p плюс 4 ст.отклонения бином.распределения
# (там, где вероятности значений почти не равны нулю)
left = int(n * p - 4 * sqrt(n * p * (1 - p)))
right = int(n * p + 4 * sqrt(n * p * (1 - p)))
 
# зададим набор значений биномиальной величины на выбранном промежутке:
# все неотрицательные целые числа от left до right
x = list(range(max(0, left), right))
# рисуем график биномиального распределения
ax = plt.subplot()
# используем синие о-образные маркеры и размер маркера 5
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=5, label='вероятности биномиального распределения')

# зададим параметры нормального распределения равными
# мат.ожиданию и дисперсии биномиального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))
 
# рисуем график нормального распределения:
# получим список из 100 чисел от left до right
x = linspace(left, right, 50)
# используем красную линию ширины 3 и прозрачности 0.3
ax.plot(x, norm.pdf(x, mu, sigma), 'o-', color='C3', lw=0.5, alpha=0.9, label='плотность вероятности нормального распределения')
 
# зададим подписи по оси x в пределах ± 3 ст.отклонений от мат.ожидания
# (последний аргумент в методе range - интервал между подписями)
x_ticks = list(range(int(mu - 3 * sigma), int(mu + 3 * sigma), 5000))
ax.set_xticks(x_ticks)

# настраиваем оптимальное расположение легенды и выводим график
ax.legend(loc='best')
plt.show()


# лежит ли наше значение 399,5 тыс в промежутке mu ± 3 sigma в промежутке от 0 до n:
if (mu - (3 * sigma) <= 399500) and ((mu + (3 * sigma) >= 399500)):
    print('Условие выполняется')
else:
    print('Условие не выполняется')
```

Это говорит о том, что кратковременными поездками пользователи тарифа `free` чаще пользуются, чем пользователи тарифа `ultra`.

Временем и расстоянием у пользователей обоих тарифа существует (слабая линейная положительная зависимость)
Это говорит о том, что когда одна из переменных увеличивается, другая переменная также имеет тенденцию к увеличению, но слабым или ненадежным образом. То есть расчитывать на увеличение времени использования за счет роста дистанций не сильно приходится. Средняя скорость по выборке значительно меньше максимально установленной документацией к электросамокатам. Это значит, что есть еще возможность наращивать дистанцию без увеличения времени использования.
